```python
from matplotlib import pyplot as plt
import pandas as pd
data=pd.read_csv('fillingarea.csv')
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']
plt.plot(ages,dev_salaries,color='orange',ls="--",label="All Devs")
plt.plot(ages,py_salaries,color='green',label='Python')
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
data=pd.read_csv('fillingarea.csv')
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']
plt.plot(ages,dev_salaries,color='orange',ls="--",label="All Devs")
plt.plot(ages,py_salaries,color='green',label='Python')
overall_median=57287
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.fill_between(ages,py_salaries,overall_median,alpha=0.5)  # overall_median is used to divide the plot in two parts. The upper and lower parts are filled within the boundary which we were given as y-axis here y-axis is py_salaries
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
data=pd.read_csv('fillingarea.csv')
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']
plt.plot(ages,dev_salaries,color='orange',ls="--",label="All Devs")
plt.plot(ages,py_salaries,color='green',label='Python')
overall_median=57287
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.fill_between(ages,py_salaries,overall_median,where=(py_salaries > overall_median),alpha=0.5) # here where is like a condition fro where we want to filled the area
plt.fill_between(ages,py_salaries,overall_median,where=(py_salaries <= overall_median),color="grey",alpha=0.5)
plt.legend()
plt.show()
```

```python
# with interpolate
from matplotlib import pyplot as plt
import pandas as pd
data=pd.read_csv('fillingarea.csv')
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']
plt.plot(ages,dev_salaries,color='black',ls="--",label="All Devs")
plt.plot(ages,py_salaries,color='red',label='Python')
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.fill_between(ages,py_salaries,dev_salaries,alpha=0.5,interpolate=True) # here where is like a condition fro where we want to filled the area
plt.legend()
plt.show()
```

```python
# with interpolate
from matplotlib import pyplot as plt
import pandas as pd
data=pd.read_csv('fillingarea.csv')
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']
plt.plot(ages,dev_salaries,color='black',ls="--",label="All Devs")
plt.plot(ages,py_salaries,color='red',label='Python')
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.fill_between(ages,py_salaries,dev_salaries,where=(py_salaries > dev_salaries),alpha=0.5,interpolate=True,label="Above Avg") # here where is like a condition fro where we want to filled the area
plt.fill_between(ages,py_salaries,dev_salaries,where=(py_salaries <= dev_salaries),color="grey",alpha=0.5,interpolate=True,label="Below Avg")
plt.legend()
plt.show()
```
