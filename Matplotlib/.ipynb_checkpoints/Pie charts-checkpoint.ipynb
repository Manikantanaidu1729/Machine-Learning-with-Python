```python
from matplotlib import pyplot as plt
slices=[30,40]
plt.pie(slices,labels=['thirty','forty'])
plt.title("My Awesome Pie Chart")
plt.show()
```

```python
from matplotlib import pyplot as plt
slices=[30,40,43,34]
labels=['A','B','C','D']
colors=['yellow','green','blue','pink']
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
slices=[59219,55466,47544,36443,35917,31991,27097,23030,30524,18523,18017,7920,7331,7201,5833]
labels=['javascript','html/css','sql','python','java','bash/shell/powershell','c#','php','c++','typescript','c','other(s)','ruby','go','assembly']
plt.style.use("fivethirtyeight")
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'white'})
plt.title("Programming languages")
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
slices=[59219,55466,47544,36443,35917]
labels=['javascript','html/css','sql','python','java']
explode=[0,0,0,0.1,0.1] # the defined part will be split outside
plt.style.use("fivethirtyeight")
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'white'},explode=explode)
plt.title("Programming languages")
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
slices=[59219,55466,47544,36443,35917]
labels=['javascript','html/css','sql','python','java']
explode=[0,0,0,0.1,0] # the defined part will be split outside
plt.style.use("fivethirtyeight")
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'white'},explode=explode,shadow=True,startangle=90)
plt.title("Programming languages")
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
slices=[59219,55466,47544,36443,35917]
labels=['javascript','html/css','sql','python','java']
plt.style.use("fivethirtyeight")
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'white','linewidth':5},shadow=True,startangle=90,autopct='%1.1f%%',radius=1.5)
plt.title("Programming languages")
plt.show()
```
