```python
import seaborn as sns
import pandas as pd
```

```python
np.random.randn(5,4)
```

```python
np.cumsum(np.random.randn(5,4))  #Return the cumulative sum of the elements along a given axis
                               # if axis is not given then the cumsum over the flattened array.
```

```python
np.cumsum(np.random.randn(5,4),axis=1)  # column wise cumsum
```

```python
np.cumsum(np.random.randn(5,4),axis=0) # row wise cumsum
```

```python
plt.style.use('classic')
x=np.linspace(0,100,10)
y=np.cumsum(np.random.randn(10,5),axis=0)  # here cumsum is cumulative sum
plt.plot(x,y)
plt.legend('ABCDE',ncol=2,loc='upper left')
```
