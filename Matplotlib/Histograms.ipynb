```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight") # it is used to start the graph at origin point
ages=[18,19,21,25,26,26,30,32,38,45,55]
bins=[10,20,30,40,50,60]  # here bins are used to calculate the how many data points in a given bin
# here 10 - 20 is a bin . In this we have 2 data points with ages 18,19.
plt.hist(ages,bins=bins,edgecolor="black")
plt.title("Ages of Respondents")
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.show()
```
