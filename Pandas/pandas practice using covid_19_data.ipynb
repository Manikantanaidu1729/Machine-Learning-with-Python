```python
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('C:/Users/Manikanta/Machine learning/covid_19_data.csv')
```

```python
df.head()
```

```python
df.drop(['SNo','Last Update'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'ObservationDate':'Date','Province/State':'Province','Country/Region':'Country'},inplace=True)
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#df.fillna('NA',inplace=True)
```

```python
imputer=SimpleImputer(strategy='constant')  # SimpleImputer provides a simple strategy to impute(fill in) missing values with either a constant value, mean , median or most frequent value of each column
```

```python
df_final=pd.DataFrame(imputer.fit_transform(df),columns=df.columns)
```

```python
df_final.info()
```

```python
df_final.head()
```
